```python
import random
import pandas as pd

# Order Size
df = pd.read_excel('PYTHON_DATA (1).xlsx', sheet_name='Dataset10')  # OrderSize_20ord
# OrderSize_1(Python)
orderSize = {}
for i, value in enumerate(df['Order size'], start=1):
    orderSize[i] = value
print(orderSize)
# Processing time
df = pd.read_excel('PYTHON_DATA (1).xlsx', sheet_name='Dataset10')  # ProcessTime_20ord
# ProcessTime_1(Python)
processTime = {}
for i, value in enumerate(df['Process time'], start=1):
    processTime[i] = value
print(processTime)
# # Due date
df = pd.read_excel('PYTHON_DATA (1).xlsx', sheet_name='Dataset10')  # DueDate_20ord
# DueDate(Python)
dueDate = {}
for i, value in enumerate(df['Duetime'].astype(int), start=1):
    dueDate[i] = value
print(dueDate)
pickingDevice = [1, 2, 3, 4, 5]
capacityPickingDevice = {1: 100, 2: 100, 3: 100, 4: 100, 5: 100}
# orderSize = {1: 27, 2: 38, 3: 1, 4: 12, 5: 23, 6: 5, 7: 12, 8: 22, 9: 36, 10:7}
# processTime = {1: 15, 2: 19, 3: 7, 4: 16, 5: 5, 6: 10, 7: 32, 8: 35, 9: 28, 10: 19}
# dueDate = {1: 30, 2: 90, 3: 95, 4: 70, 5: 25, 6: 45, 7: 29, 8: 75, 9: 55, 10: 120}
batchNumber = {}
# Make a copy order
aCopyOrder = orderSize.copy()
POPULATION_SIZE = 1000
NUMBER_OF_GENERATION = 500
SURVIVAL_RATE = 0.1
CROSSOVER_RATE = 0.85
IMMIGRATION_RATE = 0.05
MUTATION_RATE = 0.05


class Individual(object):
    def __init__(self, chromosome):
        self.chromosome = chromosome
        self.fitness = self.cal_fitness()

    @classmethod
    def create_genes(self, anOrderPool):
        # Order pool for creating genes
        orderPool = anOrderPool.copy()
        device = random.choice(pickingDevice)
        # Auto increment batch is assigned to picking device
        if device not in batchNumber:
            batch = 1
            batchNumber[device] = 1
        else:
            batch = batchNumber[device] + 1
            batchNumber[device] = batch
        # List of orders to assign a batch
        listOfOrders = []

        # Capacity of a picking device

        capacity = capacityPickingDevice[device]
        # Quantity of weight in a batch
        batchSize = 0
        # Get a list of order is randomly chosen assigned to a batch
        while batchSize < capacity:
            # Break a loop when a list pool is empty
            if not list(orderPool.keys()): break
            order = random.choice(list(orderPool.keys()))
            # Check a random order can be assigned to a batch?
            if (capacity - batchSize >= orderPool[order]):
                listOfOrders.append(order)
                batchSize += orderPool[order]
                orderPool.pop(order)
            else:
                orderPool.pop(order)
            continue
        # Remove group of orders assigned to a batch from orders pool creating chromosome
        for order in listOfOrders:
            if order in anOrderPool:
                anOrderPool.pop(order)
        gene = (device, batch, listOfOrders)

    @classmethod
    def create_gnome(self):
        global batchNumber
        global aCopyOrder
        gnome = []

        while aCopyOrder:
            gnome.append(self.create_genes(aCopyOrder))
        # Reset
        batchNumber = {}
        aCopyOrder = orderSize.copy()
        return gnome

    def create_sub_genes(self, chromo, orderPool):
        while orderPool:
            chromo.append(self.create_genes(orderPool))
        old_gnome = self.chromosome
        orderPool = {}
        # Remove two genes randomly
        for i in range(2):
            gene = random.choice(old_gnome)
            for order in gene[2]:
                orderPool[order] = orderSize[order]
            old_gnome.remove(gene)
        # Create two genes
        self.create_sub_genes(old_gnome, orderPool)
        # Update batch
        chromosome = self.update_batching(old_gnome)
        return chromosome

    def two_points(self, chromosome):
        # Generating the random two-points to perform crossover
        firstPoint = random.randrange(0, 20)
        endPoint = random.randint(firstPoint, 20)
        while firstPoint == 0 and endPoint == 20 - 1:
            firstPoint = random.randrange(0, 20)
            endPoint = random.randint(firstPoint, 20)
        return firstPoint, endPoint

    def update_batching(self, chromosome):
        batchNumber = {}
        new_chromosome = []
        for gene in chromosome:
            pickingDevice = gene[0]
            listOfOrders = gene[2]
            if pickingDevice not in batchNumber:
                batchNumber[pickingDevice] = 1
                new_chromosome.append((pickingDevice, batchNumber[pickingDevice],
                                       listOfOrders))
            else:
                batchNumber[pickingDevice] += 1
                new_chromosome.append((pickingDevice, batchNumber[pickingDevice],
                                       listOfOrders))
        return new_chromosome

    def mate(self, par2):
        partner1 = self.chromosome[:]
        partner2 = par2.chromosome[:]
        x1, x2 = self.two_points(partner1)
        y1, y2 = self.two_points(partner2)
        orders_2Points_par1 = []
        orders_2Points_par2 = []
        for gene in partner1[x1:x2 + 1]:
            listOfOrder = gene[2]
            for order in listOfOrder:
                orders_2Points_par1.append(order)
        for gene in partner2[y1:y2 + 1]:
            listOfOrder = gene[2]
            for order in listOfOrder:
                orders_2Points_par2.append(order)
        # Chromosome without two-points genes
        par1_without_TwoPoints = []
        par2_without_TwoPoints = []
        for gene in partner1:
            if gene not in partner1[x1:x2 + 1]:
                par1_without_TwoPoints.append(gene)
        for gene in partner2:
            if gene not in partner2[y1:y2 + 1]:
                par2_without_TwoPoints.append(gene)
                TwoPoints_par1 = partner1[x1:x2 + 1]

        TwoPoints_par2 = partner2[y1:y2 + 1]
        partner1[x1:x2 + 1], partner2[y1:y2 + 1] = partner2[y1:y2 + 1], partner1[x1:x2 + 1]
        partnerA = partner1[:]
        partnerB = partner2[:]

        for gene in par1_without_TwoPoints:
            if gene in TwoPoints_par2:
                partnerA.remove(gene)
            else:
                listOfOrder = gene[2]
            for order in listOfOrder:
                if order in orders_2Points_par2:
                    partnerA.remove(gene)
                    break
        for gene in par2_without_TwoPoints:
            if gene in TwoPoints_par1:
                partnerB.remove(gene)
            else:
                listOfOrder = gene[2]
                for order in listOfOrder:
                    if order in orders_2Points_par1:
                        partnerB.remove(gene)
                        break
        par1 = self.update_batching(partnerA)
        par2 = self.update_batching(partnerB)
        ordersPool_par1 = {}
        ordersPool_par2 = {}
        listOfOrders_par1 = []
        listOfOrders_par2 = []
        for gene in par1:
            listOfOrders = gene[2]
            for order in listOfOrders:
                if order not in listOfOrders_par1:
                    listOfOrders_par1.append(order)
        for gene in par2:
            listOfOrders = gene[2]
            for order in listOfOrders:
                if order not in listOfOrders_par2:
                    listOfOrders_par2.append(order)
        for order in list(orderSize.keys()):
            if order not in listOfOrders_par1:
                ordersPool_par1[order] = orderSize[order]
        for order in list(orderSize.keys()):
            if order not in listOfOrders_par2:
                ordersPool_par2[order] = orderSize[order]
        self.create_sub_genes(par1, ordersPool_par1)
        self.create_sub_genes(par2, ordersPool_par2)
        par1 = self.update_batching(par1)
        par2 = self.update_batching(par2)
        offspring = random.choice([par1, par2])
        return Individual(offspring)

    def cal_fitness(self):
        fitness = 0
        completionTimeDevice = {}
        for gene in self.chromosome:
            pickingDevice = gene[0]
            listOfOrders = gene[2]
            completionTime = 0
            # Calculate time completion of list of orders in every gene
            for order in listOfOrders:
                completionTime += processTime[order]
            if pickingDevice not in completionTimeDevice.keys():
                completionTimeDevice[pickingDevice] = completionTime
            else:
                completionTimeDevice[pickingDevice] += completionTime
                tardiness = 0
            if pickingDevice in completionTimeDevice.keys():
                completionTime = completionTimeDevice[pickingDevice]
            for order in listOfOrders:
                if dueDate[order] >= completionTime:
                    tardiness += 0
                else:
                    tardiness += completionTime - dueDate[order]
                    fitness += tardiness
        return fitness

    # def __str__(self):
    # return f"From str method of Individual: chrome is {self.chromosome}"


def main():
    global POPULATION_SIZE
    generation = 1
    population = []
    for _ in range(POPULATION_SIZE):
        gnome = Individual.create_gnome()
        population.append(Individual(gnome))
    print("Generation: {}\tFitness: {}\tString: {}\t". \
          format(generation,
                 population[0].fitness,
                 population[0].chromosome))
    # population = sorted(population, key = lambda x:x.fitness)
    # print("Generation: ", generation)
    # for i in range(POPULATION_SIZE):
    # print("Fitness: {}\tString: {}".\
    # format(population[i].fitness,
    # population[i].chromosome))
    while True:
        # sort the population in increasing order of fitness score
        population = sorted(population, key=lambda x: x.fitness)
        # if the individual having lowest fitness score ie.
        # 0 then we know that we have reached to the target
        # and break the loop

        # ------ SET STOPPING GENERATION-----#
        if generation == NUMBER_OF_GENERATION:
            break
        # if population[0].fitness <= 0:
        # found = True
        # break
        # Otherwise generate new offsprings for new generation
        new_generation = []
        # Perform Elitism, that mean 10% of fittest population
        # goes to the next generation
        s = int(POPULATION_SIZE * SURVIVAL_RATE)
        new_generation.extend(population[:s])
        # 85% of new generation come from Individuals
        # will mate to produce offspring
        s = int(POPULATION_SIZE * CROSSOVER_RATE)
        for _ in range(s):
            parent1 = random.choice(population)
            parent2 = random.choice(population)
            child = parent1.mate(parent2)
            new_generation.append(child)
        # Finally 5% go on imigration individuals
        s = int(POPULATION_SIZE * IMMIGRATION_RATE)
        for _ in range(s):
            migration = random.choice(population)
            new_generation.append(migration)
        # Mutate 5% chromosome
        s = int(POPULATION_SIZE * MUTATION_RATE)
        for _ in range(s):
            individual = random.choice(population)
            individual.chromosome = individual.mutated_genes()

        population = new_generation
        generation += 1
        print("Generation: {}\tFitness: {}\tString: {}". \
              format(generation,
                     population[0].fitness,
                     population[0].chromosome))


# population = sorted(population, key = lambda x:x.fitness)
# print("Generation: ", generation)
# for i in range(POPULATION_SIZE):
# print("Fitness: {}\tString: {}".\
# format(population[i].fitness,
# population[i].chromosome))
if __name__ == '__main__':
    main()
```

```python
import math
c=10
class Solution(object):
    def judgeSquareSum(self, c):
        temp=[]
        if math.sqrt(c)%1==0 or math.sqrt(c/2)%1==0:
            return True
        for i in range(int(math.sqrt(c-1))+1):  
            temp.append(i)
        for i in range(len(temp)):
            for j in range(i+1,len(temp)):
                if i*i+j*j==c:
                    return True
        return False
s=Solution()
s.judgeSquareSum(c)
```
